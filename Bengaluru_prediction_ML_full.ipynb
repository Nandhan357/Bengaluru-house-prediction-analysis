outlier handling

```python
def outlier_treatment(data1,columns):
  Q1=data1[columns].quantile(0.25)
  Q3=data1[columns].quantile(0.75)
  IQR=Q3-Q1
  lower_bound=Q1-1.5*IQR
  upper_bound=Q3+1.5*IQR
  return data1[(data1[columns]>=lower_bound)&(data1[columns]<=upper_bound)]
```

```python
cleaned_data=data1.copy()
```

```python
for column in numerical_data:
  cleaned_data = outlier_treatment(cleaned_data, column)
```
